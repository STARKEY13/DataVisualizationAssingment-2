```python
import requests
import networkx as nx
import matplotlib.pyplot as plt

# Function to fetch the text file from GitHub and extract the edge list
def extract_edge_list_from_github(url):
    # Make an HTTP GET request to fetch the raw file content
    response = requests.get(url)
    if response.status_code != 200:
        print("Failed to fetch the file from GitHub.")
        return None

    # Extract edge list from the text content
    edge_list = []
    for line in response.text.split('\n'):
        # Assuming each line contains an edge in the format: source_node target_node
        parts = line.strip().split()
        if len(parts) == 2:
            edge_list.append((parts[0], parts[1]))

    return edge_list

# GitHub raw file URL
github_raw_url = "https://raw.githubusercontent.com/STARKEY13/DataVisualizationAssingment-2/main/data.txt"

# Extract edge list from the GitHub file
edge_list = extract_edge_list_from_github(github_raw_url)
```

```python
G=nx.Graph()
G.add_edges_from(edge_list)

print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
layout1 = nx.spiral_layout(G)
layout2 = nx.spring_layout(G)
# Draw
nx.draw(G,layout1, with_labels=False,node_size=4,node_color='orange')
plt.title(f"Layout 1")
plt.show()


nx.draw(G,layout2, with_labels=False,node_size=4,node_color='orange')
plt.title(f"Layout 2")
plt.show()
```
